```python
import matplotlib.pyplot as plt

followers = df["followers"]/1000
following = df["following"]/500

plt.figure(figsize=(6,5))
plt.scatter(followers, following, s=70, alpha=0.8, color="teal")
plt.xlabel("Followers")
plt.ylabel("Following")
plt.title("Followers vs Following")
plt.grid(True, linestyle="-", alpha=0.4)
plt.show()
```

```python
# -----------------------------
# Blok 4: Elbow Method (Followers vs Following)
# -----------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

X = df[["followers", "following"]]
X_array = X.to_numpy()

scaler_temp = StandardScaler()
X_scaled = scaler_temp.fit_transform(X_array)

inertias = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K_range, inertias, marker='o', color='purple')
plt.xlabel("Jumlah Cluster (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method Followers vs Following")
plt.grid(True, alpha=0.3)
plt.xticks(K_range)
plt.show()

print("→ Perhatikan titik siku (elbow) pada grafik untuk menentukan K optimal.\n")
```

```python
# -----------------------------
# Blok 5: Pipeline Clustering + Visualisasi Hasil (Followers vs Following)
# -----------------------------
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('kmeans', KMeans(n_clusters=3, random_state=42, n_init=10))
])

# Fit model
pipeline.fit(X_array)

# Ambil label cluster
labels = pipeline.named_steps['kmeans'].labels_

# Tambahkan ke dataframe
df['cluster'] = labels

# Visualisasi clustering
plt.figure(figsize=(9, 6))
scatter = plt.scatter(
    df['followers'] / 1_000_000,
    df['following'] / 500,
    c=labels,
    cmap='viridis',
    s=80,
    alpha=0.85
)

plt.xlabel("Followers ")
plt.ylabel("Following")
plt.title("Hasil Clustering Akun (K=3)")
plt.colorbar(scatter, label='Cluster')
plt.grid(True, linestyle="--", alpha=0.3)
plt.show()

# Ringkasan rata-rata per cluster
print("\nRata-rata per Cluster:")
print(df.groupby('cluster')[['followers', 'following']].mean().round(0).astype(int))
```

```python
# -----------------------------
# Blok 6: Silhouette Score (tanpa warning)
# -----------------------------
from sklearn.metrics import silhouette_score
# Transform pakai numpy array yang sama
X_scaled_final = pipeline.named_steps['scaler'].transform(X_array)

sil_score = silhouette_score(X_scaled_final, labels)
print(f"\nSilhouette Score (K=3): {sil_score:.4f}")

if sil_score > 0.5:
    print("→ Cluster cukup terpisah dengan baik")
elif sil_score > 0.3:
    print("→ Cluster cukup wajar (ada sedikit overlap)")
else:
    print("→ Cluster kurang optimal, coba ubah jumlah K")
```
